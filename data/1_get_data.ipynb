## Get Entry 

The first retrieval date: 2022.11.10

The second retrieval date: 2024.01.15

```python
response = requests.get("https://plato.stanford.edu/contents.html")
print(response)
soup = BeautifulSoup(response.content, "html.parser")
l = []
for link in soup.findAll('a'):
    l.append(str(link.get('href')))
# print(len(l))
# print(l)
L = []
for s in l:
    if s.find('entries') != -1:
        L.append(s)
    
L = list(set(L))
print(len(L))
# save pcikle
with open('get_data/entry_list.pickle', 'wb') as f:
    # Pickle the 'data' dictionary using the highest protocol available.
    pickle.dump(L, f, pickle.HIGHEST_PROTOCOL)
```

```python
# sep_entry_list = []

INDEX = 0

info_dict_short_list = []
text_dict_list = []

base = 'https://plato.stanford.edu/'
for link in L:
    url = base+str(link)
    info_dict_short, text_dict = preprocessing.get_sep(url)
    # print(title)
    info_dict_short_list.append(info_dict_short)
    text_dict_list.append(text_dict)
    # text
    with open("get_data/sep_entry_text_list.json", 'w') as outfile:
        json.dump(text_dict_list, outfile, indent=4)
    # info without text
    with open("get_data/sep_entry_notext_list.json", 'w') as outfile:
        json.dump(info_dict_short_list, outfile, indent=4)
    # pickle
    with open('get_data/sep_entry_notext_list.pickle', 'wb') as f:
        pickle.dump(info_dict_short_list, f)
    
    INDEX+=1
    print(INDEX, url)
```
